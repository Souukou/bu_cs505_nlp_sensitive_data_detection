# Manually create dataset using the content generated by LLMs

We put the positive example that contains sensitive data generated by LLM in  `positive.txt`, and negative example that does not contains sensitive data in `negative.txt`.

This files is to merge those two files, strip them, and label 1 for the sensitive data, label 0 for non-sensitive data.

```python
import csv
import random

# Read in a text file and process the lines
def process_file(file_path):
    with open(file_path, 'r') as file:
        # Reading each line, stripping whitespace, removing newline characters
        # and excluding empty lines
        lines = [line.strip() for line in file.readlines() if line.strip()]

    return lines

# Write the processed lines to a CSV file
def write_to_csv(content, output_csv_path):
    with open(output_csv_path, 'w', newline='') as csvfile:
        # Creating a CSV writer object
        csvwriter = csv.writer(csvfile)

        # Writing the headers
        csvwriter.writerow(['Code', 'Label'])

        # Writing the data rows
        for line, label in content:
            csvwriter.writerow([line, label])
        

# Process the file and write to CSV
positive = process_file('positive.txt')
positive = [(x, 1) for x in positive]
negative = process_file('negative.txt')
negative = [(x, 0) for x in negative]

all_sample = positive + negative
random.shuffle(all_sample)


write_to_csv(all_sample, 'dataset.csv' )

```
